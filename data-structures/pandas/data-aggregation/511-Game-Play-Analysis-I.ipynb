## Problem Description  

You are given a table `Activity` with the following structure:  

### Table: Activity  

| Column Name  | Type  |  
|--------------|------|  
| player_id    | int  |  
| device_id    | int  |  
| event_date   | date |  
| games_played | int  |  

- `(player_id, event_date)` is the **primary key**, meaning each row represents a unique record of a playerâ€™s activity on a given date.  
- Each row shows the **activity of players** who logged in and played a number of games (possibly 0) before logging out.  
- The goal is to find the **first login date** for each player.  

### Example 1  

**Input:**  

**Activity table:**  

| player_id | device_id | event_date | games_played |  
|-----------|-----------|------------|--------------|  
| 1         | 2         | 2016-03-01 | 5            |  
| 1         | 2         | 2016-05-02 | 6            |  
| 2         | 3         | 2017-06-25 | 1            |  
| 3         | 1         | 2016-03-02 | 0            |  
| 3         | 4         | 2018-07-03 | 5            |  

**Output:**  

| player_id | first_login |  
|-----------|-------------|  
| 1         | 2016-03-01  |  
| 2         | 2017-06-25  |  
| 3         | 2016-03-02  |  

**Explanation:**  

- **Player 1:** First logged in on **2016-03-01**  
- **Player 2:** First logged in on **2017-06-25**  
- **Player 3:** First logged in on **2016-03-02**  

## Approach to Solve the Problem  

1. **Group by `player_id`**  
   - Find the **minimum `event_date`** for each `player_id`, which represents the first login date.  

2. **Rename the column to `first_login`**  
   - Ensure the output format matches the problem statement.  

3. **Return the result**  

Happy coding! ðŸŽ‰

```python
import pandas as pd

def game_analysis(activity):
    activity = activity.groupby(['player_id'])['event_date'].min().reset_index()
    activity = activity.rename(columns={'event_date': 'first_login'})
    return activity[['player_id', 'first_login']]
```

```python
# Data from LeetCode problem
data = [
    [1, 2, "2016-03-01", 5],
    [1, 2, "2016-05-02", 6],
    [2, 3, "2017-06-25", 1],
    [3, 1, "2016-03-02", 0],
    [3, 4, "2018-07-03", 5],
]

# Create the DataFrame
activity = pd.DataFrame(
    data, columns=["player_id", "device_id", "event_date", "games_played"]
).astype(
    {
        "player_id": "Int64",
        "device_id": "Int64",
        "event_date": "datetime64[ns]",
        "games_played": "Int64",
    }
)

# Call the function
game_analysis(activity)
```
