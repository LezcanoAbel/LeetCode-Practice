## Problem Description  

You are given a table `Courses` with the following structure:  

### Table: Courses  

| Column Name | Type    |  
|-------------|---------|  
| student     | varchar |  
| class       | varchar |  

- `(student, class)` is the **primary key**, meaning each student can only enroll once per class.  
- Each row indicates a student's enrollment in a particular class.  

Write a solution to **find all classes that have at least 5 students**.  

Return the result table **in any order**.  

### Example 1  

**Input:**  

**Courses table:**  

| student | class    |  
|---------|----------|  
| A       | Math     |  
| B       | English  |  
| C       | Math     |  
| D       | Biology  |  
| E       | Math     |  
| F       | Computer |  
| G       | Math     |  
| H       | Math     |  
| I       | Math     |  

**Output:**  

| class   |  
|---------|  
| Math    |  

**Explanation:**  

- **Math** has **6 students**, so it is included.  
- **English** has **1 student**, so it is **not included**.  
- **Biology** has **1 student**, so it is **not included**.  
- **Computer** has **1 student**, so it is **not included**.  

## Approach to Solve the Problem  

1. **Group by class:**  
   - Count the number of unique students in each class using `.nunique()`.  

2. **Filter Classes with at Least 5 Students:**  
   - Keep only the classes where the count of students is **greater than or equal to 5**.  

3. **Return the Result** 

Happy coding! ðŸŽ‰

```python
import pandas as pd

def find_classes(courses):
    courses = courses.groupby('class', as_index = False)['student'].nunique()
    courses = courses[courses['student'] >= 5]
    return  courses[['class']]
    
```

```python
# Data from Leetcode problem
data = [
    ["A", "Math"],
    ["B", "English"],
    ["C", "Math"],
    ["D", "Biology"],
    ["E", "Math"],
    ["F", "Computer"],
    ["G", "Math"],
    ["H", "Math"],
    ["I", "Math"],
]

# Create a DataFrame
courses = pd.DataFrame(data, columns=["student", "class"]).astype(
    {"student": "object", "class": "object"}
)

# Call the function
find_classes(courses)
```
